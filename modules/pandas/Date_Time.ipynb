```python
df['Date']
```

```python
df.loc[0, 'Date'].day_name()
```

## Convert to date format as it is being loaded

```python
# Use the date_format argument
df = pd.read_csv('data/ETH_1h.csv', parse_dates=['Date'], date_format ='%Y-%m-%d %I-%p')
```

```python
df['Date']
```

## datetime methods on series using dt class

```python
df['Date'].dt.day_name()
```

```python
# Create a new column with day of week
df['DayOfWeek'] = df['Date'].dt.day_name()
```

```python
df
```

## min() and max()

```python
df['Date'].min()
```

```python
df['Date'].max()
```

### Time Delta

```python
df['Date'].max() - df['Date'].min()
```

## Filters

```python
filt = (df['Date'] >= '2020')
df.loc[filt]
```

```python
filt = (df['Date'] >= '2019') & (df['Date'] < '2020')
df.loc[filt] 
```

```python
# Using datetime
filt = (df['Date'] >= pd.to_datetime('2019-01-01')) & (df['Date'] < pd.to_datetime('2020-01-01'))
df.loc[filt] 
```

## Re-Sampling
- Pandas date offset page: https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects

```python
df['High'].resample('D').max()
```

```python
highs = df['High'].resample('D').max()
highs.loc['2020-01-01']
```

### Plotting

```python
highs.plot();
```

## Re-Sample mutliple columns at once

```python
df.drop(columns='Symbol', inplace=True)
```

```python
df
```

```python
df.resample('W').mean()
```

## Resampling using multiple aggregation methods

```python
df.resample('W').agg({'Close': 'mean', 'High': 'max', 'Low': 'min', 'Volume': 'sum'})
```
