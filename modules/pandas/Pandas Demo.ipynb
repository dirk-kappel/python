## Casting Datatypes and Handling Missing Values

```python
import numpy as np
```

```python
people = {
    'first':['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'Missing'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df_people = pd.DataFrame(people)
```

```python
df_people
```

### Remove the rows with missing values

#### dropna()

```python
help(df_people.dropna)
```

```python
df_people.dropna()
```

```python
# Using the default arguments
df_people.dropna(axis='index', how='any')
# axis = [index, columns]
# how = Will drop row with any missing value
```

```python
# how = all, will only drop rows when all values are missing
df_people.dropna(axis='index', how='all')
```

```python
# The subset will only drop rows where email is missing
df_people.dropna(axis='index', how='any', subset=['email'])
```

```python
# Checking multiple columns to drop
# passing in how = all, both need to be missing
df_people.dropna(axis='index', how='all', subset=['last', 'email'])
# This drops rows that are missing email or last name. If it has one or the other it will not drop it.
# To make permanent you need to inplace=True
```

#### replace()
- replacing and dropping custom values

```python
help(df_people.replace)
```

```python
# Replace custom values with numpy nan
df_people = pd.DataFrame(people)

df_people.replace('NA', np.nan, inplace=True)
df_people.replace('Missing', np.nan, inplace=True)
```

```python
df_people
```

```python
df_people.dropna()
```

#### isna()
- check if value is na or not

```python
help(df_people.isna)
```

```python
df_people.isna()
```

#### fillna()
- fill na values with a string or number

```python
help(df_people.fillna)
```

```python
df_people.fillna('MISSING')
```

#### dtypes
- attribute, not a method

```python
df_people = pd.DataFrame(people)
df_people.replace('NA', np.nan, inplace=True)
df_people.replace('Missing', np.nan, inplace=True)
df_people.dtypes
```

```python
# Object means a combination of string, number, nan
df_people
```

```python
# Casting
# To get the mean we must first convert to a number
df_people['age'] = df_people['age'].astype(int)
# Missing values will cause an error.
```

```python
# Convert nan to float
df_people['age'] = df_people['age'].astype(float)
```

```python
df_people
```

```python
df_people['age'].mean()
```
