### Importing data from github url

```python
import csv
import requests
import numpy as np
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

csv_url='https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/refs/heads/master/Python/Matplotlib/02-BarCharts/data.csv'
r = requests.get(csv_url)

# Using StringIO to create a file-like object from the text content
from io import StringIO
csv_data = StringIO(r.text)
#  Read the file using the csv module
csv_reader = csv.DictReader(csv_data)
# print out first row
row = next(csv_reader)
print(row)
```

```python
# Print the languages
print(row['LanguagesWorkedWith'])
```

```python
# Split on the ;
print(row['LanguagesWorkedWith'].split(';'))
```

### Use counter class

```python
from collections import Counter
language_counter = Counter()
for row in csv_reader:
    language_counter.update(row['LanguagesWorkedWith'].split(';'))
print(language_counter)
```

### most_common() method
- Used with counter to grab the most common items in the counter

```python
print(language_counter.most_common(15))
```

```python
# Break out language and popularity into two separate lists.
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
print(languages)
print(popularity)
```

```python
# Plot the charts
plt.bar(languages, popularity)

plt.title('Most Popular Languages')
plt.xlabel('Programming Language')
plt.ylabel('Number of People Who Use')
```

### barh

```python
# Plot as a horizontal bar
plt.barh(languages, popularity)   # Using barh will plot a horizontal bar chart

plt.title('Most Popular Languages')
plt.xlabel('Number of People Who Use')
plt.show();
```

```python
# Reversing the popularity
languages_reverse=languages.copy()
popularity_reverse=popularity.copy()

languages_reverse.reverse()
popularity_reverse.reverse()

plt.barh(languages_reverse, popularity_reverse)   # Using barh will plot a horizontal bar chart

plt.title('Most Popular Languages')
plt.xlabel('Number of People Who Use')
plt.show();
```

## Using Pandas to load the Data

```python
import pandas as pd

data = pd.read_csv(csv_url)
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages_reverse=languages.copy()
popularity_reverse=popularity.copy()

languages_reverse.reverse()
popularity_reverse.reverse()

plt.barh(languages_reverse, popularity_reverse)   # Using barh will plot a horizontal bar chart

plt.title('Most Popular Languages')
plt.xlabel('Number of People Who Use')
plt.show();
```

```python
import pandas as pd
from matplotlib import pyplot as plt

data_url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/refs/heads/master/Python/Matplotlib/05-Fill_Betweens/data.csv'

data = pd.read_csv(data_url)
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287

plt.fill_between(ages, py_salaries)  # This will fill between python salaries all the way down to the bottom.
plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show();
```

### Formatting dates into datetimes and not strings

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn-v0_8')

data_url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/refs/heads/master/Python/Matplotlib/08-TimeSeries/data.csv'
data = pd.read_csv(data_url)

data['Date'] = pd.to_datetime(data['Date']) # This will convert the date from string to datetime format
data.sort_values('Date', inplace=True)  # This will sort by the dates. Not possible if it is only a string

price_date = data['Date']
price_close = data['Close']

plt.plot(price_date, price_close,
         marker='o',
         linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show();
```

## nrows, ncols

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn-v0_8')

data_url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/refs/heads/master/Python/Matplotlib/10-Subplots/data.csv'
data = pd.read_csv(data_url)
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True)  # Plot in 2 rows and 2 columns

# Access individual axes using indexing
axes[0, 0].plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

axes[0, 1].plot(ages, py_salaries, label='Python')
axes[1, 0].plot(ages, js_salaries, label='JavaScript')

axes[0, 0].legend()
axes[0, 0].set_title('Median Salary (USD) by Age')
axes[0, 0].set_ylabel('Median Salary (USD)')

axes[0, 1].legend()
axes[0, 1].set_title('Median Salary (USD) by Age')
axes[0, 1].set_xlabel('Ages')

axes[1, 0].legend()
axes[1, 0].set_title('Median Salary (USD) by Age')
axes[1, 0].set_xlabel('Ages')
axes[1, 0].set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
